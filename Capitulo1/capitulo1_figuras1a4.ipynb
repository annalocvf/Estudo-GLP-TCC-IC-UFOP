## Geração do Gráfico: Rendimento e Peso do Botijão de GLP

Este código utiliza as bibliotecas Matplotlib e NumPy para gerar um gráfico que compara o rendimento domiciliar per capita por classe de renda com o peso do preço do botijão P-13 em relação à renda média.

```python
import matplotlib.pyplot as plt
import numpy as np

classes = df_excel['Classe de Rendimento'].tolist()
rendimento = df_excel['Rendimento Médio (R$)'].tolist()
percentual = df_excel['Preço Botijão / Rendimento (%)'].tolist()

x = np.arange(len(classes))

fig, ax1 = plt.subplots(figsize=(12, 7))

ax1.bar(x, rendimento, color='skyblue', label='Rendimento domiciliar per capita')
ax1.set_ylabel("Rendimento médio (R$)", fontsize='medium')
ax1.set_xticks(x)
ax1.set_xticklabels(classes, rotation=45, ha='right')
ax1.set_xlabel("Classes de Rendimento Domiciliar Per Capita", fontsize='medium')

ax2 = ax1.twinx()
ax2.plot(x, percentual, marker='o', color='red', label='Preço do Botijão P-13 / Rendimento Médio (%)')
ax2.set_ylabel("Preço do Botijão P-13 / Rendimento Médio (%)", fontsize='medium')

for i, v in enumerate(rendimento):
    ax1.text(i, v + (max(rendimento) * 0.02), f"R$ {v}", ha='center', va='bottom')

for i, p in enumerate(percentual):
    ax2.text(i + 0.05, p + (max(percentual) * 0.02), f"{p}%", ha='left', va='bottom')

# Add a title
plt.title("Rendimento domiciliar per capita e peso do botijão P-13\nBrasil - 2022", fontsize='large')

# Add the legend
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left', bbox_to_anchor=(0.0, 1.15), frameon=False, fontsize='medium')

plt.tight_layout()
plt.show()
```

## Figura 2 – Evolução dos preços de GLP (R$ / botijão de 13 kg)

DADOS DE 2025

Fonte dos dados:  
https://www.sindigas.org.br/Download/DADOS-ESTATISTICOS/MARGEM/2025-margens-p13-tabela.pdf

Os dados utilizados neste gráfico foram extraídos da tabela “Evolução dos preços de GLP (R$ / botijão de 13 kg)”, disponibilizada pelo Sindigás com base em informações da Agência Nacional do Petróleo, Gás Natural e Biocombustíveis (ANP).

O documento apresenta a decomposição do preço do botijão P-13, mostrando os principais componentes da formação do preço ao longo do tempo.

- preço de realização do produtor  
- ICMS  
- margem bruta de distribuição  
- preço de distribuição  
- margem bruta de revenda  
- preço final ao consumidor  

As margens de distribuição e de revenda são calculadas a partir dos dados do levantamento de Preços da ANP, que monitora semanalmente os preços praticados no mercado brasileiro de combustíveis.

Como o arquivo original está em formato PDF, os dados foram extraídos e organizados manualmente em uma planilha Excel, permitindo o tratamento das informações e a geração dos gráficos utilizados nesta análise.

```python
import pandas as pd

#URL do arquivo Excel no Github
url_figura2 = 'https://github.com/annalocvf/Estudo-GLP-TCC-IC-UFOP/raw/main/Capitulo1/figura2.xlsx'


df_figura2 = pd.read_excel(url_figura2)

print("DataFrame 'df_figura2' carregado com sucesso:")
display(df_figura2.head())
```

```python
print("\nInformações gerais do DataFrame 'df_figura2':\n")
df_figura2.info()

print("\nNomes das colunas de 'df_figura2':\n")
print(df_figura2.columns)

print("\nValores únicos na coluna 'Indicador':\n")
print(df_figura2['Indicador'].unique())
```

```python
target_indicators = [
    'Margem Bruta de Revenda',
    'Margem Bruta de Distribuição',
    'Preço de Realização do Produtor',
    'ICMS',
    'Preço Final GLP P13'
]

df_figura2['Indicador'] = df_figura2['Indicador'].str.strip()

monthly_columns = [col for col in df_figura2.columns if col not in ['Indicador']]
medias_setores_atualizadas = {}

print("\nMédias mensais dos indicadores:")
for setor in target_indicators:

    current_setor = setor
    if setor == 'Preço Final GLP P13' and setor not in df_figura2['Indicador'].unique():
        if 'Preço Final ao Consumidor' in df_figura2['Indicador'].unique():
            current_setor = 'Preço Final ao Consumidor'
        else:
            print(f"  - '{setor}' ou 'Preço Final ao Consumidor' não encontrado.")
            medias_setores_atualizadas[setor] = float('nan')
            continue

    row = df_figura2[df_figura2['Indicador'] == current_setor]

    if not row.empty:
        media = row[monthly_columns].mean(axis=1).iloc[0]
        medias_setores_atualizadas[setor] = media
        print(f"  - {setor}: {media:.2f}")
    else:
        print(f"  - '{setor}' não encontrado.")
        medias_setores_atualizadas[setor] = float('nan')


print("\nDicionário de médias:")
print(medias_setores_atualizadas)
```

```python

final_price_avg = medias_setores_atualizadas.get('Preço Final GLP P13')

if pd.isna(final_price_avg) or final_price_avg == 0:
    print("Não foi possível calcular as porcentagens porque o Preço Final ao Consumidor não foi encontrado ou é zero.")
else:
    print("\nPorcentagem de cada setor no preço final ao consumidor:")
    for setor, media in medias_setores_atualizadas.items():
        if setor == 'Preço Final GLP P13':
            continue

        if pd.notna(media):
            percentage = (media / final_price_avg) * 100
            print(f"  - {setor}: {percentage:.2f}%")
        else:
            print(f"  - {setor}: Não disponível para cálculo de porcentagem (média não encontrada).")
```
